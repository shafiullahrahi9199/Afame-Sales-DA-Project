```python
# Average Customer Retention

cus=df['Customer ID'].value_counts().index
year=[]
month=[]
customer_id=[]
for i in cus:
    year.append(df[df['Customer ID']==i].sort_values(by='year').iloc[0].year)
    month.append(df[df['Customer ID']==i].sort_values(by='year').iloc[0].month)
    customer_id.append(i)
    
end_year=[]
end_month=[]
for i in cus:
    end_year.append(df[df['Customer ID']==i].sort_values(by='year').iloc[-1].year)
    end_month.append(df[df['Customer ID']==i].sort_values(by='year').iloc[-1].month)
```

```python
data=pd.DataFrame({'customer_id':customer_id,'year':year,'month':month,'end_year':end_year,'end_month':end_month})
from datetime import datetime
def days(year,month,end_year,end_month):
    start_date = datetime(year,month, 1)
    end_date = datetime(end_year, end_month, 1)
    return (end_date - start_date).days

day=[]
for i,j,k,l in zip(data.year,data.month,data.end_year,data.end_month):
    day.append(days(i,j,k,l))
```

```python
data['total_days']=day
```

```python
data
```

```python
# Average Customer Retention in Days

print(round(data.total_days.mean()),'Days')
```

```python
# Customer Retention Rate

no_retention=len(data[data.total_days==0])
retention=len(data[data.total_days>0])

plt.pie([no_retention,retention],labels=['no_retention','retention'],autopct='%1.1f%%')
```

```python
# Sales Identifier System(Program)

def sales_identify():
    country=input('Enter the Country Name: ')
    state=input('Enter the name of State: ')
    city=input('Enter City Name: ')
    sal=df[(df.Country==str(country))&(df.State==str(state))&(df.City==str(city))]
    print('Total Sales ',round(sal.Sales.sum()))
```

```python
# Avg shipping Cost by Location

def shipping_cost_identify():
    country=input('Enter the Country Name: ')
    state=input('Enter the name of State: ')
    city=input('Enter City Name: ')
    sal=df[(df.Country==str(country))&(df.State==str(state))&(df.City==str(city))]
    print('Total Shipping Cost: ',round(sal['Shipping Cost'].mean()))
    
# shipping Cost by Country

df.groupby('Country')['Shipping Cost'].mean().sort_values(ascending=False).head(40).plot(kind='bar',color=sns.color_palette("Blues"),figsize=(12,4))
```
